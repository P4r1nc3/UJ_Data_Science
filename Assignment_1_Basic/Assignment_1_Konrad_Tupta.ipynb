## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:
x = main_df[["sqft_living"]].values
y = main_df[["price"]].values
plt.scatter(x,y)
plt.xlabel("living sqft")
plt.ylabel("price")
plt.show()

#Down belown is shown relation of square footage and price.

# As can be read from the chart below, the vast majority of flats are smaller than 6,000 sqft.
# The situation is similar with the price, where most of the flats are cheaper than 3 millions.
# There are very few flats which are more expensive and larger.
```

```python
#b)
#Your code:
x1 = main_df[["grade"]].values
y1 = main_df[["price"]].values
plt.xlabel("grade")
plt.ylabel("price")
plt.scatter(x1,y1)

# Other variables correlated with price are: bathrooms, sqft_living, sqft_above, location etc...
```

```python
#c)
#Your code:
x2 = main_df[["zipcode"]].values
y2 = main_df[["price"]].values
plt.xlabel("zipcodes")
plt.ylabel("price")
plt.scatter(x2,y2)

# Here we used plt.scatter() function to show relation of zipcode and price.
# The most expensive apartment is placed in 98100 zipcode area.
```

```python
#Your code:
plt.figure(figsize=(20, 12), dpi = 80)
plt.scatter(main_df['long'], main_df['lat'], c = main_df['grade'], cmap = 'inferno')
plt.xlabel('longitude')
plt.ylabel('lattitude')
plt.set_cmap('ocean')
plt.show()

# The brighter the color is, the higher the score is. Darker color means lower score.
```

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
data = []
for i in range(1,14):
  data.append(main_df[main_df['grade'] == i]['price'])
plt.boxplot(data)
plt.ylabel("price")
plt.xlabel("grade")
plt.show() 

# In the boxplot below we can see how the price of house is speead out for the each grade.

#Compare with plot 2b):
# This boxplot looks very similar to the plot in 2b). 
# Boxplot gives more information such as median, minimum and maximum value, first and third quartile.
```

```python
#b)
#Your code:
from statsmodels.graphics.mosaicplot import mosaic

mosaic(data = main_df, index=['bedrooms', 'floors'], title='bedrooms')
plt.show()


# Mosaic Plot is a graphical method for visualizing data from two or more Qualitative data.
# It is the multidimensional extension of spineplots, which graphically display the same information 
# for only one variable.
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
houses_df = main_df.loc[(main_df['bedrooms'] == 3) & (main_df['bathrooms'] == 2) & (main_df['floors'] == 2)]
houses_df.describe()

# Here we have statisctics of apartments, which has 3 bedrooms, 2 bathrooms, and 2 floors.
```

```python
#b)
#Your code:


#Distinct accepted apartments
# Here draw scatter using the distinct accepted apartments.
plt.scatter(houses_df['grade'], houses_df['price'], s=90, c='yellow', edgecolor='black')
# Then we draw a scatter with usage main_df to show all examples of the relation between grade and price.
plt.scatter(main_df['grade'], main_df['price'], s=10, c='black')
plt.show()

```

```python
#c)
#Your code:
sort_condition = (houses_df['price'] / houses_df['sqft_living']).sort_values().index
houses_df.loc[sort_condition, :].head()

#  Below we have a list of apartments with 3 bedrooms & 2 bathrooms & 2 floors, 
#  which has the best price-to-sqft_living ratio (list is additionally sorted).
```
