```python
# I have defined below a function, that will be used to create and print charts. 

# This function takes the name of the column as a parameter and x which is the number 
# of items we want to display in our plot for example:
# if x = 10 then the chart will only include the first 10 items.
    
#Plot Function:
def plot(parameter, x):
  # Preparing datadrame 
  dataframe = pd.DataFrame(public_df[parameter].str.get_dummies(';').sum())
  dataframe.reset_index(inplace = True)
  dataframe.columns = [parameter, 'Quantity']
  dataframe = dataframe.sort_values('Quantity', ascending = False).head(x)

  # Printing dataframe
  print(dataframe)

  # Preparing plot
  plt.figure(figsize=(18, 12))
  sns.barplot(x = parameter, y = 'Quantity', data = dataframe, palette = "Blues")
  plt.xticks(rotation = 60, horizontalalignment = 'right')
```

```python
# Pie chart to show gender statistics. In my opinion, it is ideal for presenting a percentage share.
    
# This function takes the name of the column as a parameter and x which is the number 
# of items we want to display in our plot for example:
# if x = 10 then the chart will only include the first 10 items.

#Plot Function:
def pie(parameter, x):
    # Preparing datadrame 
    dataframe = public_df[parameter].value_counts().head(x)
    
    # Printing dataframe
    print(dataframe)
    
    # Preparing plot
    plt.figure(figsize = (7, 7))
    # I made my own color list, because I wanted this chart to match with the others.
    color_set = ['#e5eef6', '#d4e1ee', '#bdd3e3', '#9dc2d5', '#78aac8', '#5a94ba', '#3f7baa', '#29649a', '#386290']
    # plt.pie() is uded to plot a pie chart
    plt.pie(dataframe, labels = dataframe.index, autopct = '%1.1f%%', startangle = 180, colors = color_set);  
```
